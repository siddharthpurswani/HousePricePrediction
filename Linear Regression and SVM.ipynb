Linear regression is a statistical method used to predict a value of variable based on value of other variable.The variable you want to predict is dependent variable.The variable you are using to predict the other variable's value is called the independent variable.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("MagicBricks.csv")
```

```python
df.head()
```

```python
df = df.drop_duplicates()
```

```python
df = df.drop(['Parking','Status'],axis= "columns")
```

```python
df
```

```python
df.isnull().sum()
```

```python
df
    
    
```

```python
df.dropna(subset = ["Bathroom","Furnishing","Per_Sqft"], inplace = True)
```

```python
df.head()
```

```python
df["Locality"].unique()
```

```python
df["Locality"] = df["Locality"].apply(lambda x : x.split(",")[-1])
```

```python
df = df[df['Locality'].str.len() < 50]
```

```python
df["Locality"] = df["Locality"].apply(lambda x: x.strip())
```

```python
df["Locality"] = df["Locality"].apply(lambda x: x.split()[0:2])
```

```python
df["Locality"] = df["Locality"].apply(lambda x: ",".join(x))
```

```python
df["Locality"] = df["Locality"].apply(lambda x: x.replace(","," "))
```

```python
df
```

```python
df.sort_values(["Per_Sqft"],ascending = False).head(10)
```

```python
df.head()
```

```python
df["Per_Sqft"].describe()
```

```python
Q1 = df["Per_Sqft"].quantile(0.25)
Q3 = df["Per_Sqft"].quantile(0.75)
IQR = Q3 - Q1
lower_limit = Q1 - 1.5 * IQR
upper_limit = Q3 + 1.5 * IQR
df_new = df[(df["Per_Sqft"] >= lower_limit) & (df["Per_Sqft"] <= upper_limit)]
```

```python
df_new
```

```python
df_new = df_new[~(df_new.Area/df_new.BHK<300)]
df_new
```

```python
df_new = df_new[((df_new.Area*df_new.Per_Sqft)*0.7) <= (df_new.Price)]
```

```python
location_dummies = pd.get_dummies(df_new.Locality,drop_first=True)
Furnishing_dummies = pd.get_dummies(df_new.Furnishing,drop_first=True)
Transaction_dummies = pd.get_dummies(df_new.Transaction,drop_first=True)
```

```python
df_new = pd.concat([df_new,location_dummies,Furnishing_dummies,Transaction_dummies],axis ="columns")
```

```python
df_new["Price_log"] = np.log(df_new["Price"])
```

```python
df_new
```

```python
df_new = df_new.drop(["Furnishing","Locality","Transaction","Type"],axis = "columns")
```

```python
x = df_new.drop(["Price","Price_log"],axis= "columns")
x
```

```python
y = df_new["Price_log"]
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.4)
```

```python
from sklearn.linear_model import LinearRegression
lrReg = LinearRegression()
lrReg.fit(x_train,y_train)
lrReg.score(x_test,y_test)
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score
```

```python
cv = ShuffleSplit(n_splits=5,test_size=0.4, random_state= 0)
cross_val_score(LinearRegression(),x,y,cv=cv)
```

```python
np.exp(lrReg.predict(x_test)[2])
```
